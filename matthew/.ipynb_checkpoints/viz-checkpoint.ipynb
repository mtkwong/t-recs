```python
import numpy as np
import pandas as pd
import time as tm
import datetime as dt
import matplotlib.pyplot as plt
from collections import defaultdict
import csv
import ast
import json

svdPath = 'comparison/svd.csv'
df_svd = pd.read_csv(svdPath, sep=',')

knnbuPath = 'comparison/knn_basic_ub.csv'
df_knnbu = pd.read_csv(knnbuPath, sep=',')

knnbiPath = 'comparison/knn_basic_ib.csv'
df_knnbi = pd.read_csv(knnbiPath, sep=',')

knnmuPath = 'comparison/knn_means_ub.csv'
df_knnmu = pd.read_csv(knnbuPath, sep=',')

knnmiPath = 'comparison/knn_means_ib.csv'
df_knnmi = pd.read_csv(knnbuPath, sep=',')

knnzuPath = 'comparison/knn_zscore_ub.csv'
df_knnzu = pd.read_csv(knnbuPath, sep=',')

knnziPath = 'comparison/knn_zscore_ib.csv'
df_knnzi = pd.read_csv(knnbuPath, sep=',')

ccPath = 'comparison/co_clustering.csv'
df_cc = pd.read_csv(ccPath, sep=',')

print('done')
```

```python
fig, axs = plt.subplots(4,2,figsize=(15,15))

x_svd, y_svd = df_svd['rmse'][:25].apply(lambda x:(x-1)*1000), np.arange(25)
axs[0,0].barh(y=y_svd, width=x_svd, color='blue')
axs[0,0].invert_yaxis()
axs[0,0].set_title('SVD')
axs[0,0].set_xlabel('RMSE')

# TODO
axs[0,0].annotate(s='lr_bu={}, lr_bi={}, lr_pu={}, lr_qi={},\nreg_bu={}, reg_bi={}, reg_pu={}, reg_qi={}'.format(df_svd['lr_bu'][0],
                  df_svd['lr_bi'][0],df_svd['lr_pu'][0],df_svd['lr_qi'][0],df_svd['reg_bu'][0],df_svd['reg_bi'][0],df_svd['reg_pu'][0],df_svd['reg_qi'][0]),
                  xy=(0,0),xytext=(0,0))
#axs[0,0].set_yticklabels(None)

x_cc, y_cc, k_cc, m_cc = df_cc['rmse'], df_cc['mae'], df_cc['k'], df_cc['min_k']
axs[0,1].scatter(x=x_cc, y=y_cc, alpha=0.3)
axs[0,1].set_title('Co-Clustering')
axs[0,1].set_xlabel('RMSE')
axs[0,1].set_ylabel('MAE')
for i in (0,9,10):
    axs[0,1].annotate(s='({},{})'.format(k_cc[i],m_cc[i]),xy=(x_cc[i],y_cc[i]),xytext=(x_cc[i]-0.0012,y_cc[i]-0.005))

plt.show()

print('done')
```
