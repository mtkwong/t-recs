```python
def findType(categories):
    df_f = df_reviews[df_reviews['categaries'] == categories]\
           .groupby(['type'])\
           .count()\
           .sort_values(['categaries'], ascending = False)
    df_f['type'] = df_f.index
    print(df_f[:5])
    df_f = df_f.reset_index(drop = True)
    return df_f.loc[0].type
```
